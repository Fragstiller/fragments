```python
%matplotlib inline

import pickle as pkl
from typing import cast
from fragments.params import ParamStorage, ParamCell
from fragments.strategy import ConditionalStrategy, ConditionLogic, Strategy, Action
from fragments.indicators import RSI, SMA, Indicator
from fragments.optim import optimize
from fragments.stats import equity, sqn, plot
```

```python
with open("../data/BTCUSDT.pkl", "rb") as f:
    ohlcv_list = pkl.load(f)[500000:]
Indicator.enable_precalculation(ohlcv_list)
Strategy.set_fee(0.022)
```

```python
ParamStorage.new_global()
conditional_1 = ConditionalStrategy(RSI())
conditional_1.on_condition.bounds = [Action.BUY]
conditional_2 = ConditionalStrategy(RSI(), previous=conditional_1)
conditional_2.on_condition.bounds = [Action.SELL]
cast(ParamCell, conditional_2.condition_logic).bounds = [ConditionLogic.IGNORE]
optimize(conditional_2, equity, ohlcv_list, random_state=42, verbose=True, initial_point_generator="sobol")
```
